```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(drop='first')
encoder.fit(df[['salary']]) #must be in 2D array or df

encoded_embarked=encoder.transform(df[['salary']])
print(encoder.categories_)
encoded_embarked.toarray()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(drop='first')
encoder.fit(df[['Department']]) #must be in 2D array or df

encoded_embarked=encoder.transform(df[['Department']])
print(encoder.categories_)
encoded_embarked.toarray()
```
